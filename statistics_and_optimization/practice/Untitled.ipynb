## Variance and Standard Deviation

```python
import numpy as np
import pandas as pd
```

```python
Alice = np.array([4, 5, 6, 5, 5, 4, 6])
Bob = np.array([2, 10, 5, 8, 1, 3, 9])
```

```python
print(np.mean(Alice))
# Alice’s variance is small (small squared differences from the mean), meaning her walking distances are consistent.
print(np.var(Alice))
print(np.mean(Bob))
# Bob’s variance is large, meaning he has big variations in his walking routine.
print(np.var(Bob))

# If Alice’s standard deviation is 1 km, it means on most days, she walks about 1 km away from her average distance.
print(np.std(Alice))
# If Bob’s standard deviation is 4 km, it means he walks around 4 km more or less than his average on any given day.
print(np.std(Bob))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Step 1: Define probability of success (clicking the email)
p_click = 0.3  # 30% of users click the email

# Step 2: Simulate 1000 users receiving the email
n_users = 1000
clicks = bernoulli.rvs(p_click, size=n_users)  # Generates 0s (no click) and 1s (click)

# Step 3: Count results
click_count = np.sum(clicks)
no_click_count = n_users - click_count

print(f"Total Users: {n_users}")
print(f"Users who clicked: {click_count}")
print(f"Users who did not click: {no_click_count}")

# Step 4: Plot the results
plt.bar(["No Click (0)", "Click (1)"], [no_click_count, click_count], color=['red', 'green'])
plt.xlabel("Outcome")
plt.ylabel("Number of Users")
plt.title("Bernoulli Distribution: Email Click Rate Simulation")
plt.show()
```
