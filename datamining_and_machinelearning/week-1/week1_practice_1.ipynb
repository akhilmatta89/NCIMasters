## checking for null values

```python
df.isnull().sum()
```

```python
desc = ["Quantity","Unit Price","Total Sales"]
sales = [df.qty.sum(),df.unit_price.sum(),df.total.sum()]

plt.bar(desc,sales)
plt.title('Coffee Sales')
plt.xlabel('Data')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.subplot(2, 2, 1)
sns.histplot(df['qty'], kde=True)
plt.title('Distribution of Quantity')
plt.subplot(2, 2, 2)
sns.histplot(df['unit_price'], kde=True)
plt.title('Distribution of Unit Price')
plt.subplot(2, 2, 3)
sns.histplot(df['total'], kde=True)
plt.title('Distribution of Total Price')
plt.tight_layout()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df['date'], df['qty'],
label='Quantity')
plt.plot(df['date'], df['unit_price'],
label='Unit Price')
plt.plot(df['date'], df['total'],
label='Total Price')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series of CoLee Sales')
plt.legend()
plt.show()
```
