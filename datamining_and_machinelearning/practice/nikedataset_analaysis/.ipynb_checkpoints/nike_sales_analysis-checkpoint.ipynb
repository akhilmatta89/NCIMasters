#### General Data Exploration:
    - What are the top-selling regions and months based on revenue or units sold?
    - How do different product lines perform across various regions?
    - Are there any seasonal trends in sales? Do certain months see higher sales than others?

```python
# Group by Region and sum the Units_Sold
df_grouped = df.groupby("Region", as_index=False)["Units_Sold"].sum()

# Sort by Units Sold for better visualization
df_grouped = df_grouped.sort_values("Units_Sold", ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 5))
bars = plt.bar(df_grouped["Region"], df_grouped["Units_Sold"], color="skyblue")

# Annotate bars with values
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f"{height:,}", ha="center", va="bottom", fontsize=10)

# Labels and title
plt.xlabel("Region")
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold per Region")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

from the above graph we can see china has most of the units sold and Southeast ranks last in the list

```python
df_months_revenue_groued = df.groupby("Month", as_index=False)["Revenue_USD"].sum()
df_months_revenue_groued
```

```python
plt.figure(figsize=(15, 5))
bars = plt.bar(df_months_revenue_groued["Month"], df_months_revenue_groued["Revenue_USD"])
plt.title("Month vs Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue_USD")
# Annotate bars with values
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f"{height:,}", ha="center", va="bottom", fontsize=10)
```

Revenue wise if we see december is having lot of revenue and April is having less revenue
    (Might be because of christmas and new year eve)

```python
plt.figure(figsize=(15, 8))  # Increase width and height
sns.barplot(data=df, x="Month", y="Revenue_USD", hue="Region", errorbar=None)
plt.xlabel("Month", fontsize=14)  # Increase label size
plt.ylabel("Revenue (USD)", fontsize=14)
plt.title("Revenue by Month and Region", fontsize=16)
plt.xticks(rotation=45, fontsize=12)  # Adjust tick labels
plt.yticks(fontsize=12)
plt.legend(title="Region", fontsize=12)
plt.show()
```

```python
# finding outliers using IQR technique
q1 = (25/100)*len(df["Units_Sold"])
q3 = (75/100)*len(df["Units_Sold"])
print(q1, q3)
print(df["Units_Sold"][250])
print(df["Units_Sold"][750])
```

```python
Q1 = df["Units_Sold"].quantile(0.25)  # 25th percentile
Q3 = df["Units_Sold"].quantile(0.75)  # 75th percentile
print(Q1, Q3)
```

```python
lower_fence = Q1 - 1.5*(Q3 - Q1)
upper_fence = Q1 + 1.5*(Q3 - Q1)
print(lower_fence, upper_fence)
```

```python
# Initialize the label encoder
encoder = LabelEncoder()

# Apply label encoding
df.loc[:, 'Month_Encoded'] = encoder.fit_transform(df['Month'])
df.loc[:, 'Region_Encoded'] = encoder.fit_transform(df['Region'])
df.loc[:, 'Main_Category_Encoded'] = encoder.fit_transform(df['Main_Category'])
df.loc[:, 'Sub_Category_Encoded'] = encoder.fit_transform(df['Sub_Category'])
df.loc[:, 'Product_Line_Encoded'] = encoder.fit_transform(df['Product_Line'])
df.loc[:, 'Price_Tier_Encoded'] = encoder.fit_transform(df['Price_Tier'])
```

```python
df.drop(columns=['Month', 'Region', 'Main_Category','Sub_Category','Product_Line','Price_Tier'], inplace=True)
df.rename(columns={'Month_Encoded': 'Month'}, inplace=True)
df.rename(columns={'Region_Encoded': 'Region'}, inplace=True)
df.rename(columns={'Main_Category_Encoded': 'Main_Category'}, inplace=True)
df.rename(columns={'Sub_Category_Encoded': 'Sub_Category'}, inplace=True)
df.rename(columns={'Product_Line_Encoded': 'Product_Line'}, inplace=True)
df.rename(columns={'Price_Tier_Encoded': 'Price_Tier'}, inplace=True)
```
