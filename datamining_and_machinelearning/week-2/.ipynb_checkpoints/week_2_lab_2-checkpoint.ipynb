### Drop rows with null values

### Filling the missing values with mean approach

```python
df.loc[:, 'RenewableEnergy'] = df['RenewableEnergy'].fillna(df['RenewableEnergy'].mean())
df.loc[:, 'Temperature'] = df['Temperature'].fillna(df['Temperature'].mean())
df.loc[:, 'Humidity'] = df['Humidity'].fillna(df['Humidity'].mean())
```

### Display all duplicate rows

```python
duplicates = df[df.duplicated()]
print(duplicates)
```

### Remove duplicate rows

```python
df = df.drop_duplicates()
```

### Label Encoding for HVACUsage, LightingUsage, Holiday

```python
# Initialize the label encoder
encoder = LabelEncoder()

# Apply label encoding
df.loc[:, 'HVACUsage_Encoded'] = encoder.fit_transform(df['HVACUsage'])
df.loc[:, 'LightingUsage_Encoded'] = encoder.fit_transform(df['LightingUsage'])
df.loc[:, 'Holiday_Encoded'] = encoder.fit_transform(df['Holiday'])
```

```python
# droppiung columns HVACUsage, LightingUsage, Holiday columns and renaming encoding columns to intial names
df.drop(columns=['HVACUsage', 'LightingUsage', 'Holiday'], inplace=True)
df.rename(columns={'HVACUsage_Encoded': 'HVACUsage'}, inplace=True)
df.rename(columns={'LightingUsage_Encoded': 'LightingUsage'}, inplace=True)
df.rename(columns={'Holiday_Encoded': 'Holiday'}, inplace=True)
```

### Encode DayOfWeek column with one-host encoding

```python
df['DayOfWeek_duplicate'] = df[["DayOfWeek"]]
df = pd.get_dummies(df, columns=["DayOfWeek_duplicate"], prefix='day_of_week')

df.loc[:, 'DayOfWeek_Encoded'] = encoder.fit_transform(df['DayOfWeek'])
df.drop(columns=['DayOfWeek'], inplace=True)
df.rename(columns={'DayOfWeek_Encoded': 'DayOfWeek'}, inplace=True)
```

### Split your dataset into X with the columns (
        - Timestamp 
        - Temperature 
        - Humidity
        - SquareFootage 
        - Occupancy 
        - HVACUsage 
        - LightingUsage
        - RenewableEnergy 
        - DayOfWeek 
        - Holiday 
### Y with the column
        - EnergyConsumption

```python
# Converting the 'Timestamp' column to datetime
# Convert 'Timestamp' column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d-%m-%y %H:%M')

# Convert Timestamp to a numerical format (UNIX timestamp in seconds)
df['Timestamp'] = df['Timestamp'].astype('int64') // 10**9  # Convert to seconds

# Define feature set X and target variable y
X = df[['Timestamp', 'Temperature', 'Humidity', 'SquareFootage', 'Occupancy', 
        'HVACUsage', 'LightingUsage', 'RenewableEnergy', 'Holiday']]
y = df['EnergyConsumption']
```

### Normalize all the data in the X with the exception of the Timestamp column.

```python
# Drop the 'Timestamp' column
X_no_timestamp = X.drop(columns=['Timestamp'])
feature_names = X_no_timestamp.columns
# Apply MinMaxScaler to the remaining features
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_no_timestamp), columns=X_no_timestamp.columns)
```

### Perform feature selection using RandomForest Classifier to use only top 8 features from the X.

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Perform Time Series Split
n_splits = 5  # Adjust as needed
tscv = TimeSeriesSplit(n_splits=n_splits)

# Initialize the model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
# Training & Evaluation
for train_index, test_index in tscv.split(X_scaled):
    X_train, X_test = X_scaled.iloc[train_index], X_scaled.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Train model
    rf.fit(X_train, y_train)


# Get feature importances
feature_importances = rf.feature_importances_

# Sort features based on importance
sorted_features = sorted(zip(feature_names, feature_importances), key=lambda x: x[1])

# Print the sorted features
for feature, score in sorted_features:
    print(f"{feature}: {score}")
```

### Visulaization of top features

```python
features, scores = zip(*sorted_features)  
# Plot bar graph
plt.figure(figsize=(8, 5))
plt.bar(features, scores, color='skyblue')  # Vertical bar chart
plt.xlabel("Features")
plt.ylabel("Feature Importance Score")
plt.title("Feature Importance in Ascending Order")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
