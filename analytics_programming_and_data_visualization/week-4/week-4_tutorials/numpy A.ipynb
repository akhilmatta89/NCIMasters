### b) Use the ravel function to flatten the data. What’s the difference between ravel and flatten?

_order_

'C': row major (default. 'F': column major 'A': flatten in column-major order, if a is Fortran contiguous in memory, row-major order otherwise 'K': flatten a in the order the elements occur in the memory

```python
#The ravel() function is used to create a contiguous flattened array.
# is a library-level function and hence can be called on any object that can successfully be parsed.
#Ravel is faster than flatten() as it does not occupy any memory.
np.ravel(b, "C")
```

```python
np.ravel(b, "F")
```

```python
#The flatten() function is used to get a copy of an given array collapsed into one dimension.
# a method of an ndarray object and hence can only be called for true numpy arrays
b.flatten()
```

```python
b
```

flatten is a method of an ndarray object and hence can only be called from numpy arrays - it always returns a copy.
ravel is a library-level function - it returns a view - ravel will often be faster since no memory is copied
